```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = '/content/Accidental_Drug_Related_Deaths.csv'  # Adjust path if needed
df = pd.read_csv(file_path)

# Drop NaN values in 'Age' column as there are 2 Missing values in 'Age' column
ages = df['Age'].dropna()

# Create histogram
bins = 20  # Number of bins
counts, bin_edges = np.histogram(ages, bins=bins)

# Compute midpoints of bins
bin_midpoints = (bin_edges[:-1] + bin_edges[1:]) / 2

# Plot histogram with light green color
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=bins, color='lightgreen', edgecolor='black', alpha=0.7, label="Histogram")

# Plot line connecting midpoints
plt.plot(bin_midpoints, counts, marker='o', linestyle='-', color='red', label="Midpoint Line")

# Labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Frequency Distribution')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
# prompt: ok visualize it into 2 chars one for deaths distribution by year and another for death distribution by month . remember the date format in 'Date' column is MM/DD/YYYY.  for both charts i want each and every bar in the x-axis to be labelled so like show each and every month and each and every year. good for months replace the month name by number because the original dataset shows month number .  now generate me two tables for this along with the two charts you gave me earlier

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Accidental_Drug_Related_Deaths.csv")

# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Extract year and month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Deaths distribution by year
deaths_by_year = df.groupby('Year')['Date'].count()

plt.figure(figsize=(12, 6))
plt.bar(deaths_by_year.index, deaths_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.title('Deaths Distribution by Year')
plt.xticks(deaths_by_year.index, rotation=45)
plt.tight_layout()
plt.show()

# Create a table for deaths by year
year_table = pd.DataFrame({'Year': deaths_by_year.index, 'Deaths': deaths_by_year.values})
print("Deaths by Year Table:")
print(year_table)


# Deaths distribution by month
deaths_by_month = df.groupby('Month')['Date'].count()

plt.figure(figsize=(12, 6))
plt.bar(deaths_by_month.index, deaths_by_month.values)
plt.xlabel('Month')
plt.ylabel('Number of Deaths')
plt.title('Deaths Distribution by Month')
plt.xticks(deaths_by_month.index)
plt.tight_layout()
plt.show()

# Create a table for deaths by month
month_table = pd.DataFrame({'Month': deaths_by_month.index, 'Deaths': deaths_by_month.values})
print("\nDeaths by Month Table:")
month_table
```

**'Race' column**

Now moving to the 'Race' column we first print the unique values to check what are they then we will decide what to do next

```python
# Unique and Misisng values of Race column
# Count occurrences of each race
race_counts = df['Race'].value_counts()
print("Unique values in 'Race' column and their counts:\n", race_counts)

# Count and print the number of missing values
missing_race_count = df['Race'].isnull().sum()
print(f"\nNumber of missing values in 'Race' column: {missing_race_count}")
```

This data is from an American context so it will better if we classified the Races they way government agencies do it there. They recognize 5 RACES namely:  American Indian or Alaska Native, Asian, White, Native Hawaiian or Other Pacific Islander, and Black or African American . We will use the 6th category: 'Unknown/Unreported' for both null and for 'Other (Specify)'. For Ethnicity there are two choices 'Hispanic or Latino' and 'Not Hispanic or Latino'. We will now have to perform appropriate replacements.




```python
# Data Mapping Races
import pandas as pd

# Replace values in the 'Race' column
df['Race'] = df['Race'].replace(['White', 'white'], 'White')
df['Race'] = df['Race'].replace(['Black or African American', 'Black'], 'Black or African American')
df['Race'] = df['Race'].replace(['Asian Indian', 'Asian, Other', 'Other Asian', 'Asian/Indian', 'Asian', 'Chinese', 'Other Asian (Specify)', 'Korean'], 'Asian')
df['Race'] = df['Race'].replace(['American Indian or Alaska Native', 'Native American, Other'], 'American Indian or Alaska Native')
df['Race'] = df['Race'].replace('Hawaiian', 'Native Hawaiian or Other Pacific Islander')
df['Race'] = df['Race'].replace(['Unknown', 'Other', 'Other (Specify)'], 'Unknown/Unreported')
df['Race'] = df['Race'].fillna('Unknown/Unreported')

# Handle specific 'Other (Specify)' cases
df.loc[df['Race'] == 'Other (Specify) portugese, Cape Verdean', 'Race'] = 'White'
df.loc[df['Race'] == 'Other (Specify) Puerto Rican', 'Race'] = 'White'
df.loc[df['Race'] == 'Other (Specify) Haitian', 'Race'] = 'Black or African American'


# Print unique values and their counts
race_counts = df['Race'].value_counts()
race_counts
```

Notice we have a rather interesting case where the person identifies as both 'Black or African American / American Indian Lenni Lenape'. The latter counts as 'American Indian or Alaska Native'. I will count this person as 'Black or African American' simply because it is the first part of their self-identified dual itentity

```python
# Data Mapping that one case
df['Race'] = df['Race'].replace('Black or African American / American Indian Lenni Lenape', 'Black or African American')

# Print unique values and their counts
race_counts = df['Race'].value_counts()
race_counts
```

NIH data only counts 2 ethnicities namely 'Hispanic or Latino' and 'Not Hispanic or Latino'. We will also add a Third option 'Unknown' for missing values. Chicanos are Mexican-origin Americans. Government surveys count people with Mexican origin as 'Hispanic or Latino' regardless of ethnic background whereas Chicanos usually self-identify as 'Chicano or Mexican American'. In this study I have , in accordance with the government's way, counted them as 'Hispanic or Latino'.

```python
# Code for data mapping in 'Ethnicity' column

# Replace values in the 'Ethnicity' column
df['Ethnicity'] = df['Ethnicity'].replace(['No, not Spanish/Hispanic/Latino', 'Not Spanish/Hispanic/Latino'], 'Not Hispanic or Latino')
df['Ethnicity'] = df['Ethnicity'].replace(['Yes, Mexican, Mexican American, Chicano',
                                           'Mexican, Mexican American, Chicano',
                                           'Spanish/Hispanic/Latino', 'Puerto Rican',
                                           'Yes, Other Spanish/Hispanic/Latino (Specify)',
                                           'Other Spanish/Hispanic/Latino', 'Yes, other Spanish/Hispanic/Latino',
                                           'Yes, Puerto Rican', 'Cuban'], 'Hispanic or Latino')
df['Ethnicity'] = df['Ethnicity'].replace('Hispanic', 'Hispanic or Latino')
df['Ethnicity'] = df['Ethnicity'].fillna('Unknown') #Fill NaN values with 'Unknown'

# Print unique values and their counts
ethnicity_counts = df['Ethnicity'].value_counts()
ethnicity_counts
```

# Fixing the Capitalization issue with City names in 'InjuryCityGeo' and 'ResidenceCityGeo' columns

We have an issue though. In InjuryCityGeo and ResidenceCityGeo the enitre city name is capitalized. We only want the first letter to be capitalized.

```python
# prompt: load dataset 'CheckPoint_5.csv'

import pandas as pd

# Load the dataset
df = pd.read_csv('CheckPoint_5.csv')

```

```python
# 'ResidenceCityGeo' Capitalization fix

import pandas as pd

def title_case_city(city_string):
    if pd.isnull(city_string):
        return city_string
    parts = city_string.split(',')
    if len(parts) > 0:
        city_part = parts[0].strip()
        words = city_part.split()
        title_cased_words = [word.capitalize() for word in words]
        title_cased_city = ' '.join(title_cased_words)
        if len(parts) > 1:
            return title_cased_city + ',' + ','.join(parts[1:])
        else:
          return title_cased_city
    return city_string


df['ResidenceCityGeo'] = df['ResidenceCityGeo'].apply(title_case_city)
print(df['ResidenceCityGeo'].head(10))
```

```python
# 'InjuryCityGeo' Capitalization fix

import pandas as pd
def title_case_city(city_string):
    if pd.isnull(city_string):
        return city_string
    parts = city_string.split(',')
    if len(parts) > 0:
        city_part = parts[0].strip()
        words = city_part.split()
        title_cased_words = [word.capitalize() for word in words]
        title_cased_city = ' '.join(title_cased_words)
        if len(parts) > 1:
            return title_cased_city + ',' + ','.join(parts[1:])
        else:
          return title_cased_city
    return city_string

df['InjuryCityGeo'] = df['InjuryCityGeo'].apply(title_case_city)
print(df['InjuryCityGeo'].head(10))
```

```python
# Mapping code for 'Description of Injury'
import pandas as pd

# Load the dataset
df = pd.read_csv('RAW2.csv')

# Define a function to categorize injury descriptions
def categorize_injury(description):
    if pd.isnull(description):
        return 'Unspecified/Miscellaneous'
    description = str(description).lower()  # Convert to lowercase for case-insensitive matching
    if any(keyword in description for keyword in ['alcohol', 'ethanol']):
        return 'Alcohol Abuse'
    elif any(keyword in description for keyword in ['medication', 'medications', 'prescribed', 'prescription']):
        return 'Medicine Overdose'
    elif any(keyword in description for keyword in ['drug', 'substance', 'acute']):
        return 'Drug Abuse'
    else:
        return 'Unspecified/Miscellaneous'

# Apply the categorization function to the 'Description of Injury' column
df['Description of Injury'] = df['Description of Injury'].apply(categorize_injury)

# Display the counts of each category
injury_category_counts = df['Description of Injury'].value_counts()
injury_category_counts
```

```python
# prompt: load FINAL.csv dataset then encode column 'Sex and then display first 10 rows so i can see what u have done. do mention what value has been encoded by what number. After that do the same for 'Race' column , 'Injury Place' column, 'Description of Injury' column, 'Death Place' column, 'Heroin' , 'Cocaine', 'Fentanyl', 'Fentanyl Analogue', 'Oxycodone', 'Oxymorphone', 'Ethanol', 'Hydrocodone', 'Benzodiazepine', 'Methadone', 'Meth/Amphetamine', 'Amphet', 'Tramad', 'Hydromorphone', 'Morphine (Not Heroin)', 'Xylazine', 'Gabapentin', 'Opiate NOS', 'Heroin/Morph/Codeine', 'Any Other Opioid' , and 'Date Type' columns also. fill the nan values of 'Other' and 'Other Significant Conditions' with 'N/A'. Then save as csv titled Encoded2

import pandas as pd

# Load the dataset
df = pd.read_csv('FINAL.csv')

# Define a function for encoding
def encode_column(df, column_name):
    unique_values = df[column_name].unique()
    mapping = {value: i for i, value in enumerate(unique_values)}
    print(f"Encoding for '{column_name}': {mapping}")
    df[column_name] = df[column_name].map(mapping)
    return df

# List of columns to encode
columns_to_encode = ['Sex', 'Race', 'Injury Place', 'Description of Injury',
                     'Death Place', 'Heroin', 'Cocaine', 'Fentanyl',
                     'Fentanyl Analogue', 'Oxycodone', 'Oxymorphone', 'Ethanol',
                     'Hydrocodone', 'Benzodiazepine', 'Methadone', 'Meth/Amphetamine',
                     'Amphet', 'Tramad', 'Hydromorphone', 'Morphine (Not Heroin)',
                     'Xylazine', 'Gabapentin', 'Opiate NOS', 'Heroin/Morph/Codeine',
                     'Any Other Opioid', 'Date Type']

# Encode the specified columns
for column in columns_to_encode:
    df = encode_column(df, column)

# Display first 10 rows
print(df.head(10))

# Save the encoded DataFrame to a new CSV file
df.to_csv('Encoded2.csv', index=False)
```

Now we have a new problem 'Other' and 'Other Significant Conditions' columns have become null again. I will fix it by imputing null with 'N/A.

```python
# prompt: load Encoded.csv and display count of nan values in 'Other' and 'Other Significant Features' columns then replace all nan values in both columns with 'N/A' and display count of nan values in btoh columns after the  replacement. save as csv titled FinalCleanDataset

import pandas as pd

# Load the dataset
df = pd.read_csv('Encoded2.csv')

# Count NaN values in 'Other' and 'Other Significant Features' columns
nan_count_other_before = df['Other'].isnull().sum()
nan_count_other_sig_before = df['Other Significant Conditions'].isnull().sum()

print(f"NaN values in 'Other' before replacement: {nan_count_other_before}")
print(f"NaN values in 'Other Significant Conditions' before replacement: {nan_count_other_sig_before}")

# Replace NaN values with 'N/A'
df['Other'].fillna('N/A', inplace=True)
df['Other Significant Conditions'].fillna('N/A', inplace=True)

# Count NaN values after replacement
nan_count_other_after = df['Other'].isnull().sum()
nan_count_other_sig_after = df['Other Significant Conditions'].isnull().sum()

print(f"NaN values in 'Other' after replacement: {nan_count_other_after}")
print(f"NaN values in 'Other Significant Conditions' after replacement: {nan_count_other_sig_after}")

# Save the DataFrame to a new CSV file
df.to_csv('FinalCleanDataset.csv', index=False)
```

Substances use visualized

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv('FinalCleanDataset.csv')

# Corrected substance-related columns based on actual dataset names
substance_columns = [
    "Heroin", "Cocaine", "Fentanyl", "Fentanyl Analogue", "Oxycodone", "Oxymorphone",
    "Ethanol", "Hydrocodone", "Benzodiazepine", "Methadone", "Meth/Amphetamine", "Amphet",
    "Tramad", "Hydromorphone", "Morphine (Not Heroin)", "Xylazine", "Gabapentin",
    "Opiate NOS", "Heroin/Morph/Codeine", "Any Other Opioid", "Other"
]

# Convert substance columns to numeric, coercing errors to NaN
for col in substance_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

#Fill NaN values with 0
df[substance_columns] = df[substance_columns].fillna(0)

# Count occurrences of each substance being used (value = 1)
substance_counts = df[substance_columns].sum().sort_values(ascending=False)

# Plot the substance usage counts
plt.figure(figsize=(12, 6))
sns.barplot(x=substance_counts.index, y=substance_counts.values, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Substances")
plt.ylabel("Usage Count")
plt.title("Usage Count of Different Substances")
plt.grid(axis="y")

# Show the plot
plt.show()
```

```python
# Compute the correlation matrix for substance-related columns
substance_corr = df[substance_columns].corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(substance_corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Co-Occurring Drugs")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)

# Show the plot
plt.show()
```
